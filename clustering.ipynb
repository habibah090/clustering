1. What is unsupervised learning in the context of machine learning
- Unsupervised learning is a type of machine learning where the model is trained on data without labeled outcomes. In other words, the algorithm tries to find patterns, structures, or relationships in the input data without being told what the correct outputs should be.

2. How does K-Means clustering algorithm work
- how the K-Means clustering algorithm works:

Choose the number of clusters
ùêæ
K.

Initialize centroids: Randomly select
ùêæ
K points from the data as initial centroids.

Assign points to clusters: For each data point, assign it to the nearest centroid (using Euclidean distance).

Update centroids: Recalculate each centroid as the mean of all points assigned to that cluster.

Repeat steps 3 and 4 until the centroids do not change significantly (convergence) or a maximum number of iterations is reached.

3. Explain the concept of a dendrogram in hierarchical clustering
- A dendrogram is a tree-like diagram used to visualize the results of hierarchical clustering.

Key Concepts:

Each leaf of the tree represents an individual data point.

Branches show how clusters are merged step by step.

The height of the branches (vertical axis) represents the distance or dissimilarity between clusters when they are joined.

Cutting the dendrogram at a specific height gives a desired number of clusters.

4. What is the main difference between K-Means and Hierarchical Clustering
- K-Means requires you to predefine the number of clusters (K).

Hierarchical Clustering does not require specifying the number of clusters upfront; it builds a tree (dendrogram) and you can choose clusters by cutting the tree at a desired level.

5. What are the advantages of DBSCAN over K-Means
- Advantages of DBSCAN over K-Means:

No need to specify the number of clusters in advance.

Can find clusters of arbitrary shapes, unlike K-Means which assumes spherical clusters.

Handles noise and outliers effectively (labels them as noise).

Does not assume equal cluster sizes or densities, unlike K-Means.

6. When would you use Silhouette Score in clustering
- You use the Silhouette Score in clustering to:

Evaluate clustering quality ‚Äî how well each point fits within its cluster.

Compare different clustering algorithms or settings (e.g., different values of K).

Determine the optimal number of clusters by choosing the value that gives the highest average silhouette score.

7.  What are the limitations of Hierarchical Clustering
- ### **Limitations of Hierarchical Clustering:**

- **Scalability**:

   * Computationally expensive for large datasets (time complexity is (O(n^2))).

- **No correction**:

   * Once a merge or split is done, it **cannot be undone**.

- **Sensitive to noise and outliers**:

   * Can produce poor results if the data contains noise.

- **Choice of linkage affects results**:

   * Different linkage methods (single, complete, average) can give **different cluster structures**.

- **No objective way to choose number of clusters**:

   * Requires manual inspection (e.g., cutting the dendrogram).

- Why Feature Scaling is Important in K-Means:

Ensures equal contribution:

 10.  Define inertia in the context of K-Means
 - Inertia in K-Means is the sum of squared distances between each data point and the centroid of the cluster it belongs to.

It measures how tight the clusters are.

Lower inertia means data points are closer to their centroids, indicating better clustering.


11. What is the elbow method in K-Means clustering

Run K-Means for different values of
ùêæ

Calculate the inertia (sum of squared distances to centroids) for each

Plot
ùêæ

Look for the ‚Äúelbow‚Äù point where the rate of decrease in inertia sharply slows down.

The elbow point suggests the best trade-off between cluster quality and number of clusters.

13. Can hierarchical clustering be used on categorical data
- Yes, hierarchical clustering can be used on categorical data, but it requires:

Using an appropriate distance or similarity measure designed for categorical variables (e.g., Hamming distance, Jaccard similarity).

Choosing a suitable linkage method to build clusters.

Since traditional Euclidean distance doesn‚Äôt work well with categorical data, you must pick or define a distance metric that handles categories.

14. What does a negative Silhouette Score indicate
- A negative Silhouette Score indicates that a data point is likely misclassified ‚Äî it is closer to points in another cluster than to points in its own cluster.

15. Explain the term "linkage criteria" in hierarchical clustering
- Linkage criteria in hierarchical clustering define how the distance between two clusters is calculated when merging them.

Common types:

Single linkage: Distance between the closest points in the two clusters.

Complete linkage: Distance between the farthest points in the two clusters.

Average linkage: Average distance between all pairs of points in the two clusters.

Ward‚Äôs linkage: Minimizes the total within-cluster variance after merging.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities

- K-Means can perform poorly on data with varying cluster sizes or densities because:

Assumes clusters are spherical and similar size:
It tries to minimize variance around centroids, which works best if clusters have roughly equal size and density.

Sensitive to outliers and uneven densities:
Clusters with higher density or larger size can dominate the centroid calculation, causing smaller or less dense clusters to be misclassified or merged incorrectly.

Equal weighting of points:
Points in dense clusters pull centroids more strongly, biasing the clustering.

17. What are the core parameters in DBSCAN, and how do they influence clustering
- The core parameters in **DBSCAN** are:

1. **Epsilon (Œµ)**

   * Defines the radius of the neighborhood around a point.
   * Influences how close points need to be to be considered neighbors.
   * Smaller Œµ ‚Üí more clusters (or more noise), larger Œµ ‚Üí fewer clusters (may merge clusters).

2. **MinPts (Minimum Points)**

   * Minimum number of points required within Œµ-radius for a point to be a **core point**.
   * Controls the minimum density to form a cluster.
   * Larger MinPts ‚Üí stricter definition of dense regions (fewer clusters, more noise).

### Influence on Clustering:

* **Œµ** affects cluster size and shape sensitivity.
* **MinPts** affects cluster density requirements and noise identification.
* Together, they control how clusters are detected and which points are labeled as noise.

19. What is agglomerative clustering
- Agglomerative clustering is a type of hierarchical clustering that builds clusters bottom-up.

How it works:

Start with each data point as its own individual cluster.

At each step, merge the two closest clusters based on a chosen distance metric and linkage criteria.

Repeat until all points are merged into a single cluster or until a stopping condition is met (e.g., desired number of clusters).

It produces a dendrogram showing the hierarchy of merges.

20. What makes Silhouette Score a better metric than just inertia for model evaluation
- Silhouette Score is often considered better than inertia for clustering evaluation because:

1. **Considers both cohesion and separation**:

   * Silhouette measures how close points are to their own cluster (cohesion) **and** how far they are from neighboring clusters (separation).
   * Inertia only measures cohesion (within-cluster variance).

2. **Works well for comparing different numbers of clusters**:

   * Silhouette Score helps identify the natural cluster structure by balancing compactness and separation.
   * Inertia always decreases as you increase clusters, which can be misleading.

3. **Scale-independent and interpretable**:

   * Silhouette Score ranges from -1 to 1, with clear interpretation.
   * Inertia values depend on data scale and are harder to interpret directly.

So, Silhouette Score provides a more holistic and interpretable evaluation of clustering quality.

```python
'''
23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot
'''

import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np

# Generate synthetic moon-shaped data
X, y_true = make_moons(n_samples=300, noise=0.05, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Plotting
plt.figure(figsize=(8, 6))

# Core points and clusters
unique_labels = set(labels)
colors = plt.cm.get_cmap('viridis', len(unique_labels))

for label in unique_labels:
    if label == -1:
        # Noise points (outliers) in red
        color = 'red'
        marker = 'x'
        size = 50
    else:
        color = colors(label)
        marker = 'o'
        size = 50
    class_member_mask = (labels == label)
    plt.scatter(X[class_member_mask, 0], X[class_member_mask, 1],
                c=[color], marker=marker, s=size, label=f'Cluster {label}' if label != -1 else 'Noise')

plt.title("DBSCAN Clustering on make_moons Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
'''
29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart
'''

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

sil_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(score)

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(k_values, sil_scores, color='skyblue')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for different values of k')
plt.xticks(k_values)
plt.show()
```

```python
'''
31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries
'''

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Generate synthetic data with overlapping clusters
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Create mesh grid for decision boundaries
h = 0.1  # step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict cluster for each point in mesh grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries and clusters
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap=cmap_bold, edgecolor='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='yellow', marker='X', label='Centroids')
plt.title('K-Means Clustering with Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
'''
43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
side-by-side
'''

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate noisy circular data
X, y_true = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

# Standardize data (important for DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Plotting side-by-side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# KMeans result
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', s=30)
axes[0].set_title('KMeans Clustering')

# DBSCAN result
axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='viridis', s=30)
axes[1].set_title('DBSCAN Clustering')

for ax in axes:
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```

```python
'''
44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering
'''

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load and scale the Iris dataset
iris = load_iris()
X = iris.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Compute silhouette scores for each sample
silhouette_vals = silhouette_samples(X_scaled, cluster_labels)

# Plot silhouette scores
plt.figure(figsize=(8, 6))
y_lower = 10
for i in np.unique(cluster_labels):
    cluster_silhouette_vals = silhouette_vals[cluster_labels == i]
    cluster_silhouette_vals.sort()
    size_cluster_i = cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, cluster_silhouette_vals,
                      alpha=0.7)

    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, f'Cluster {i}')
    y_lower = y_upper + 10  # space between clusters

plt.axvline(x=np.mean(silhouette_vals), color="red", linestyle="--")
plt.xlabel('Silhouette Coefficient')
plt.ylabel('Samples')
plt.title('Silhouette Plot for KMeans on Iris Dataset')
plt.show()
```

```python
'''
46.  Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features)
'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the Wine dataset
wine = load_wine()
X = wine.data[:, :4]  # Select the first 4 features
feature_names = wine.feature_names[:4]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Create a DataFrame for Seaborn
df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = labels.astype(str)  # Convert to string for better legend

# Create a pairplot
sns.pairplot(df, hue='Cluster', palette='tab10', diag_kind='hist')
plt.suptitle('KMeans Clustering on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```
